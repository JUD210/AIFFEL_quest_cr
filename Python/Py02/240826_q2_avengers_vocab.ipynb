````python
# 입력받은 텍스트를 전처리 후 반환한다.
def process_text(s):
  s = s.lower()
  s = s.replace('.', '')
  s = s.replace('?', '')
  s = s.replace('!', '')
  s = s.replace(':', '')
  s = s.replace(';', '')
  s = s.replace('(', '')
  s = s.replace(')', '')
  s = s.replace('"', '')
  s = s.replace("'", '')
  s = s.replace('-', '')
  s = s.replace('\n', ' ')
  return s

# 파일에 있는 글자를 불러와서 문자열을 반환한다.
def file_to_str(file_path):
  with open(file_path, 'r') as file:
    txt = process_text(file.read())
  return txt


### 1. 텍스트 전처리를 수행한다.
txt_list = file_to_str('Avengers.txt').split()


### 2. 단어 별 빈도수를 딕셔너리 형태로 저장한다.
def make_dict_from_word_count(txt_list):
  word_counts = {}
  for word in txt_list:
    # 'a' 가 있으면 1개 늘리고
    if word in word_counts:
      word_counts[word] += 1

    # 'a' 가 없으면 1개를 넣는다.
    else:
      word_counts[word] = 1
      # word가 'a'인 경우
      # == word_counts.update('a', 1)

  return word_counts

word_counts = make_dict_from_word_count(txt_list)
print(f"word_counts: {word_counts}")

### 3-1. 딕셔너리를 빈도수 순으로 내림차순 정렬한다.
## 딕셔너리의 키와 값을 각각 별도의 리스트로 저장
def sort_dict_descending_by_value(d):
  keys = list(d.keys())
  values = list(d.values())

  ## 값(벨류) 기준으로 내림차순 정렬 (버블 정렬 사용)
  # 만약 리스트의 길이가 5이면 (ex | 3 1 2 4 7)
  """
  # 맨 왼쪽부터 오른쪽으로 4번 바꾸고
  3-1 2 4 7
  3 1-2 4 7
  3 2 1-4 7
  3 2 4 1-7
  3 2 4 7 1

  # 맨 왼쪽부터 오른쪽으로 3번 바꾸고
  3-2 4 7 1
  3 2-4 7 1
  3 4 2-7 1
  3 4 7 2 1

  # 맨 왼쪽부터 오른쪽으로 2번 바꾸고
  3-4 7 2 1
  4 3-7 2 1
  4 7 3 2 1

  # 맨 왼쪽부터 오른쪽으로 1번 바꾼다
  4-7 3 2 1
  7 4 3 2 1
  """
  for i in range(len(values)):
    for j in range(i + 1, len(values)):
      if values[i] < values[j]:  # 값을 비교하여 정렬
          # 값이 더 크면 자리 바꿈 (값뿐만 아니라 키도 함께 교환)
          values[i], values[j] = values[j], values[i]
          keys[i], keys[j] = keys[j], keys[i]

  ## 정렬된 키와 값으로 새로운 딕셔너리 생성
  # ex) 'a': 7, 'the': 4, 'to': 3, 'moon': 1
  sorted_d = {}
  for i in range(len(keys)):
    sorted_d[keys[i]] = values[i]

  return sorted_d


sorted_dict = sort_dict_descending_by_value(word_counts)
print(f"sorted_dict: {sorted_dict}")

### 3-2. 정렬 순서대로 정수 인덱스를 부여한다.
def add_idx_to_dict(d):
  # ex) 'a': 0, 'the': 1, 'to': 2, 'moon': 3
  for idx, key in enumerate(d):
    d[key] = idx

  return d

indexed_dict = add_idx_to_dict(sorted_dict)
print(f"indexed_dict: {indexed_dict}")

### 4. 텍스트를 input()으로 입력받아서 정수를 return하는 함수를 만든다.
def search_dict(indexed_dict):
  txt = process_text(input("문장을 입력해주세요: "))
  txt_list = txt.split()
  # ex) to the moon
  # ex) [2, 1, 3]

  answer = []
  for word in txt_list:
    if word in indexed_dict:
      answer.append(indexed_dict[word])
  return answer

print(search_dict(indexed_dict))
````

def sort_dict_by_value(d):
    # 딕셔너리의 키와 값을 각각 별도의 리스트로 저장
    keys = list(d.keys())
    values = list(d.values())

    # 값(벨류) 기준으로 정렬 (버블 정렬 사용)
    for i in range(len(values)):
        for j in range(i + 1, len(values)):
            if values[i] < values[j]:  # 값을 비교하여 정렬
                # 값이 더 크면 자리 바꿈 (값뿐만 아니라 키도 함께 교환)
                values[i], values[j] = values[j], values[i]
                keys[i], keys[j] = keys[j], keys[i]

    # 정렬된 키와 값으로 새로운 딕셔너리 생성
    sorted_dict = {}
    for i in range(len(keys)):
        sorted_dict[keys[i]] = values[i]

    return sorted_dict

# 사용 예시
word_counts = {'hello': 5, 'world': 2, 'python': 8}

print(sorted_word_counts)  # 출력: {'python': 8, 'hello': 5, 'world': 2}
