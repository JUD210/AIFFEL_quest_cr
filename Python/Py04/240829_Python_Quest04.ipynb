````python
from math import pow

# -> 는 리턴 타입 힌트로, 무결성을 보증하기 위해 넣었음.
def input_valid_num(num: int = 0, oper: str = "") -> int:
    """
    유효한 정수를 입력받습니다. 필요 시 ZeroDivisionError를 확인합니다.

    :param num: 초기값 0
    :param oper: 연산자(빈 문자열이 기본값)
    :return: 유효한 정수
    """
    if oper == "":
        order = "첫 번째"
    else:
        order = "두 번째"

    while True:
        try:
            # 사용자가 입력한 값을 정수로 변환
            num = int(input(f"{order} 정수를 입력해주세요: "))

            # ZeroDivisionError 체크하기 (나눗셈 연산 시 0으로 나누는 경우)
            if oper == "/" and num == 0:
                raise ZeroDivisionError

        except ValueError:
            # 사용자가 정수를 입력하지 않은 경우
            print("잘못된 입력입니다. 정수를 입력해주세요.")
        except ZeroDivisionError:
            # 나눗셈 연산 시 두 번째 정수가 0인 경우
            print("0으로 나눌 수 없습니다.")
        else:
            break

    return num

def input_valid_oper() -> str:
    """
    유효한 연산자를 입력받습니다.

    :return: 유효한 연산자
    """
    while True:
        oper = input("연산자(+, -, /, *, **) 중 1개를 입력해주세요: ")

        # 지원하지 않는 연산자를 입력한 경우
        if oper not in ["+", "-", "/", "*", "**"]:
            print("연산자(+, -, /, *, **) 중 1개를 입력해주세요.")
        else:
            break

    return oper

def print_result(first_num: int, second_num: int, oper: str) -> None:
    """
    입력받은 연산자에 따라 두 정수의 연산 결과를 출력합니다.

    :param first_num: 첫 번째 정수
    :param second_num: 두 번째 정수
    :param oper: 연산자
    :return: None
    """
    if oper == "+":
        result = first_num + second_num
    elif oper == "-":
        result = first_num - second_num
    elif oper == "/":
        # 나눗셈 연산 수행
        result = first_num / second_num
    elif oper == "*":
        # 곱셈 연산 수행
        result = first_num * second_num
    else:
        # 제곱 연산 수행
        result = pow(first_num, second_num)

    print(f"{first_num} {oper} {second_num} = {result}")

def ask_continue() -> bool:
    """
    사용자가 계산을 계속할지 묻고, 프로그램 종료 여부를 반환합니다.

    :return: 계산 계속 여부(True/False)
    """
    while True:
        continue_calc = input("계속 계산하시겠습니까? (y/n): ").lower()

        # 사용자가 계산을 종료하려는 경우
        if continue_calc == "n":
            print("프로그램이 종료되었습니다.")
            return False
        # 사용자가 계산을 계속하려는 경우
        elif continue_calc == "y":
            return True
        else:
            # 유효하지 않은 입력에 대해 다시 입력 요청
            print("y 또는 n을 입력해주세요.")

# 메인 루프
main_power = True
while main_power:
    # 적절한 첫 번째 정수 입력받기
    first_num = input_valid_num(num=0)

    # 적절한 연산자 입력받기
    oper = input_valid_oper()

    # 적절한 두 번째 정수 입력받기
    second_num = input_valid_num(num=0, oper=oper)

    # 연산 결과 출력
    print_result(first_num, second_num, oper)

    # 연산 계속 할까요? (y/n)
    main_power = ask_continue()
````
